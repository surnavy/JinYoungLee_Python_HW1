```python
def add_data(friend,count):
    
    Pos=-1
    
    for i in range(len(katok)):
        
        pair=katok[i]
        
        if count>=pair[1]:
            Pos=i
            break
            
    if Pos == -1:
        Pos=len(katok)
        
    insert_data(Pos,(friend,count))
    
def insert_data(position,friend):
    if position<0 or position>len(katok):
        print("   .")
        return
    
    katok.append(None)
    kLen=len(katok)
    
    for i in range(kLen-1,position,-1):
        katok[i]=katok[i-1]
        katok[i-1]=None
        
    katok[position]=friend
    
def delete_data(position):
    if position<0 or position>len(katok):
        print("   .")
        return
    
    kLen=len(katok)
    katok[position-1]=None
    
    for i in range(position,kLen):
        katok[i-1]=katok[i]
        katok[i]=None
        
    del(katok[kLen-1])
    
katok=[]
select=-1

if __name__=="__main__":
    while(select!=3):
        
        select=int(input("(1:, 2:, 3:)-->"))
        
        if (select==1):
            data = eval(input(" -->"))
            list(data)
            add_data(data[0],data[1])
            print(katok)
            
        elif(select==2):
            pos=int(input(" -->"))
            delete_data(pos)
            print(katok)
            
        elif(select==3):
            exit
            
        else:
            print("1~3  .")
            continue
```
